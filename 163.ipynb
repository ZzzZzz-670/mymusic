```python
# 新增函数：获取热门歌曲
def get_artist_top_songs(artist_id, limit=10):
    """获取歌手的10首热门歌曲"""
    url = f"https://music.163.com/api/artist/top/song"
    params = {
        "id": artist_id,
        "limit": limit,
        "offset": 0
    }
    
    time.sleep(random.uniform(0.5, 1.5))
    
    response = safe_request(url, params)
    if response:
        try:
            data = response.json()
            return data.get("songs", [])[:limit]  # 返回前10首热门歌曲
        except json.JSONDecodeError:
            print("歌曲API返回数据解析失败")
    return []

def get_song_lyric(song_id):
    """获取歌曲歌词并去除时间轴"""
    url = f"https://music.163.com/api/song/lyric"
    params = {
        "id": song_id,
        "lv": -1  # 获取完整歌词
    }
    
    # 添加延迟避免被封
    time.sleep(random.uniform(0.5, 1.5))
    
    response = safe_request(url, params)
    if response:
        try:
            data = response.json()
            lyric = data.get('lrc', {}).get('lyric', '')
            # 去除时间轴
            clean_lyric = re.sub(r'\[.*?\]', '', lyric).strip()
            return clean_lyric
        except json.JSONDecodeError:
            print(f"歌词API返回数据解析失败 - 歌曲ID: {song_id}")
    return ""

def get_song_comments(song_id, limit=3):
    """获取歌曲评论（最多3条）"""
    url = f"https://music.163.com/api/v1/resource/comments/R_SO_4_{song_id}"
    params = {
        "limit": limit,
        "offset": 0
    }
    
    # 添加延迟避免被封
    time.sleep(random.uniform(0.5, 1.5))
    
    response = safe_request(url, params)
    if response:
        try:
            data = response.json()
            comments = []
            for comment in data.get('comments', [])[:limit]:
                comments.append({
                    "content": comment.get('content', ''),
                    "time": time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(comment.get('time', 0)/1000))
                })
            return comments
        except json.JSONDecodeError:
            print(f"评论API返回数据解析失败 - 歌曲ID: {song_id}")
    return []

# 新增函数：保存歌曲信息
def save_songs_to_csv(artist_id, artist_name, songs, filename="songs.csv"):
    """保存歌曲信息到CSV文件"""
    with open(filename, mode='a', newline='', encoding='utf-8') as file:  # 使用追加模式
        writer = csv.writer(file)
        # 如果是新文件，写入标题行
        if file.tell() == 0:
            writer.writerow([
                "歌曲ID", "歌曲名", "歌手名", "歌词", "歌曲图片URL", 
                "歌曲链接", "评论1内容", "评论1时间", "评论2内容", "评论2时间", "评论3内容", "评论3时间"
            ])
        
        for i, song in enumerate(songs):
            # 获取歌曲详细信息
            song_id = song.get('id', '')
            song_name = song.get('name', '')
            pic_url = song.get('al', {}).get('picUrl', '') if song.get('al') else ''
            song_url = f"https://music.163.com/song?id={song_id}"
            
            # 获取歌词
            lyric = get_song_lyric(song_id)
            
            # 获取评论
            comments = get_song_comments(song_id, 3)
            
            # 写入行
            writer.writerow([
                song_id,
                song_name,
                artist_name,
                lyric,
                pic_url,
                song_url,
                comments[0]['content'] if len(comments) > 0 else '',
                comments[0]['time'] if len(comments) > 0 else '',
                comments[1]['content'] if len(comments) > 1 else '',
                comments[1]['time'] if len(comments) > 1 else '',
                comments[2]['content'] if len(comments) > 2 else '',
                comments[2]['time'] if len(comments) > 2 else '',
            ])
            
            print(f"已保存歌曲: {song_name} (歌手: {artist_name})")
            
            # 减速反爬
            if (i + 1) % 3 == 0:
                wait_time = random.uniform(0.5, 1)
                print(f"已处理 {i+1}/{len(songs)} 首歌曲, 暂停 {wait_time:.1f} 秒...")
                time.sleep(wait_time)
            else:
                time.sleep(random.uniform(0.2, 0.5))

with open("songs.csv", "w", encoding="utf-8") as f:
        pass  # 清空文件
if artists:    
    # 获取每位歌手的热门歌曲并保存
    for artist in artists[:10]:  # 只爬取前10位歌手避免被封
        artist_id = artist['id']
        artist_name = artist['name']
        
        print(f"\n开始获取歌手 {artist_name} 的热门歌曲...")
        songs = get_artist_top_songs(artist_id, 10)  # 获取10首热门歌曲
        
        if songs:
            save_songs_to_csv(artist_id, artist_name, songs)
            print(f"成功保存 {len(songs)} 首歌曲")
        else:
            print(f"无法获取歌手 {artist_name} 的歌曲")
        
        # 爬完一位歌手后暂停
        wait_time = random.uniform(2, 5)
        print(f"完成歌手 {artist_name} 的爬取, 暂停 {wait_time:.1f} 秒...")
        time.sleep(wait_time)
    
    print("\n所有歌手和歌曲信息爬取完成！")
```
